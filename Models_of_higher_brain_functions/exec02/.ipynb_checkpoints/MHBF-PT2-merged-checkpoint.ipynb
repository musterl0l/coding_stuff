```python
import numpy as np
import scipy.linalg as sp
import math
import matplotlib.pyplot as plt
from helper import *
```

```python
# function that creates a random signal with specified spectrum
def random_sig(T,epsilon):
    rand_sig = np.random.randn(T)
    return gaussian_spectrum_1D(rand_sig, epsilon)
```

```python
T = 1000     #number of samples
tau_1 = 30
tau_2 = 100
# compute two signals with different epsilon
signal_1 = random_sig(T,1/math.pow(tau_1,2))
signal_2 = random_sig(T,1/math.pow(tau_2,2))
```

```python
# normalize signals
signal_1 = signal_1 - signal_1.mean()
signal_2 = signal_2 - signal_2.mean()

# plot signals
plt.plot(signal_1,label='tau = 30')
plt.plot(signal_2, label='tau = 100')
plt.legend()
plt.title("Two random signals")
plt.show()

print("A higher tau leads to a signal with higher frequencies")
```

```python
# combine the vectors in a matrix
X = np.zeros((2,1000))
X[0] = signal_1
X[1] = signal_2
```

```python
# compute covariance matrix on the data
C = np.cov(X)
# compute temporal derivative of the covariance matrix
C_diff = np.dot(np.diff(X),np.diff(X).T)/(T-1)
```

```python
# compute eigenvecors and eigenvalues
ev = sp.eig(C_diff,C)

ind = np.real(ev[0]).argsort()
ev1 = np.real(ev[1][ind[0]])
ev2 = np.real(ev[1][ind[1]])
```

```python
# normalize eigenvectors
ev1 = ev1/np.linalg.norm(ev1)
ev2 = ev2/np.linalg.norm(ev2)
```

```python
print("Extracted eigenvectors (sorted):")
print(ev1)
print(ev2)
```

For SFA, I would expect the slower signal to have the highest contribution. I would expect a higher contribution from the second signal, which is in the second column of matrix X.
In the extracted eigenvectors, it can indeed be seen that the 'weights' of the slowest signal are almost [0,1], meaning almost all of the slowest signal originates from the second signal.

```python
def plot_ev_for_all_T(tau_1, tau_2):
    eigen_vectors = np.zeros((20,2,2))
    j = 0

    for T in np.logspace(1,4,20):
        # compute two signals with different epsilon
        signal_1 = random_sig(int(T),1/math.pow(tau_1,2))
        signal_2 = random_sig(int(T),1/math.pow(tau_2,2))

        # normalize signals
        signal_1 = signal_1 - signal_1.mean()
        signal_2 = signal_2 - signal_2.mean()

        # combine the vectors in a matrix
        X = np.zeros((2,int(T)))
        X[0] = signal_1
        X[1] = signal_2

        # compute covariance matrix on the data
        C = np.cov(X)
        # compute temporal derivative of the covariance matrix
        C_diff = np.dot(np.diff(X),np.diff(X).T)/(T-1)

        # compute eigenvecors and eigenvalues
        ev = sp.eig(C_diff,C)

        ind = np.real(ev[0]).argsort()
        ev1 = np.real(ev[1][ind[0]])
        ev2 = np.real(ev[1][ind[1]])

        # normalize eigenvectors
        ev1 = ev1/np.linalg.norm(ev1)
        ev2 = ev2/np.linalg.norm(ev2)

        # store them
        eigen_vectors[j,0] = ev1
        eigen_vectors[j,1] = ev2
        j = j + 1
        
    plt.plot(np.logspace(1,4,20),np.abs(eigen_vectors[:,0,0]),label='ev1[0]')
    plt.plot(np.logspace(1,4,20),np.abs(eigen_vectors[:,0,1]),label='ev1[1]')
    plt.plot(np.logspace(1,4,20),np.abs(eigen_vectors[:,1,0]),label='ev2[0]')
    plt.plot(np.logspace(1,4,20),np.abs(eigen_vectors[:,1,1]),label='ev2[1]')
    plt.legend()
    plt.semilogx()
    plt.xlabel('T [# of samples]')
    plt.show()
```

```python
plt.title("dtau = 50")
plot_ev_for_all_T(50, 100)
plt.title("dtau = 150")
plot_ev_for_all_T(50, 200)
plt.title("dtau = 250")
plot_ev_for_all_T(50, 300)
plt.title("dtau = 0")
plot_ev_for_all_T(50, 50)
```

#### How does the SFA-mixture change depending on the length of the input signal? Can you give an intuition why?
As the number of samples in the original signal is growing the eigenvectors are getting more and more orthogonal in the empirical calculations, meaning that the SFA can seperate the fast and slow wave better for higher input sample sizes. Therefore I suggest using longer samples for a good SFA.

The SFA is better at seperating the fast and slow signal for longer signals as for short signals. This is due to the randomness that is introduced which may result in the faster signal having a sinosoidal shape on which the faster signal is superimposed, which the SFA interprets as a slow wave. As the sample size grows larger, these random features are more attenuated, as the time window for the SFA is bigger and can seperate random fluctuations from actual waves.

#### How do the eigenvalues depend on the τ ’s that you chose?
The larger the difference between tau of the two signals, the less samples have to be used for the eigenvectors to converge. This is because a larger dtau results in a higher difference of 'slowness' between the two input signals, which makes it easier to separate. If dtau is zero, any combinations of the two signals can lead to the slowest signal, as they are equally slow.

```python
T_im = 10**4
eps_im = 0.04

im = random_sig(T_im, eps_im)
im2d = im.reshape(100,100)
```

```python
def return_RF_input_matrix(image,win_L):
    # Being sure the image is flattened:
    im = image.flatten()
    inp_matrix = np.zeros([win_L, len(im)-win_L])
    for i in range(len(im)-win_L):
        inp_matrix[:,i] = im[i:i+win_L]
    return inp_matrix
```

```python
X_im = return_RF_input_matrix(im,100)
```

```python
X_im.mean(axis=1)
```

```python
def normalize_RF_matrix(RF_inp):
    # iterating through each row and normalize them to 1
    for i in range(np.shape(RF_inp)[0]):
        norm_factor = np.mean(RF_inp[i]) - 1
        RF_inp[i] -= norm_factor
    return RF_inp
```

```python
norm_X_im = normalize_RF_matrix(X_im)
```

```python
# Compute C and Cdiff from the normalized RF input matrix 
C_im = np.cov(norm_X_im)
C_diff_im = np.dot(np.diff(norm_X_im),np.diff(norm_X_im).T)/(T_im-1)
```

```python
# compute eigenvecors and eigenvalues
eigvals, eigvect = sp.eig(C_diff_im,C_im)
```

```python
# first n smallest eigvalue
n = 5
# eigvalue indices
#idx = np.argpartition(eigvals, n)
idx = eigvals.argsort()[:n]

print('The first {} slowest eigenvectors are:\n'.format(n))

fig, ax = plt.subplots(1,n,figsize=[n*5,5])
for i, ax_ in enumerate(ax):
    plt.axes(ax_)
    plt.axis('off')
    plt.title('#{}'.format(idx[i]))
    plt.plot(eigvect[:,idx[i]])

plt.show()
```
