##### (1)

dW(t) is a Wiener process (stochastic part)
dW here does not depend on time

now:
function f is mu = const
function f is sigma = const

```python
sigma_1 = 0.5
mu_1 = 0.2
x_0 = 0
dt = 0.01
ts = np.arange (0,10,dt)
it = 5


def mu(y,t):
    return mu_1
def sigma(y,t):
    return sigma_1

def dW(delta_t): 
    return np.random.normal(loc = 0.0, scale = np.sqrt(delta_t))

ys = np.zeros(ts.shape)
ys[0] = x_0

for _ in range (it):
    for i in range (1,ts.size):
        t = (i-1)*dt
        y = ys[i-1]
        ys[i] = y + mu(y,t) * dt + sigma(y,t) * dW(dt)
    plt.plot(ts,ys)
    
plt.show()
```

##### (2)

one solution: simulate trails unitl you have 200 that ended in 10 seconds
other solution: simulate longer than 10 seconds (e.g. 20s)

decision boundary +/- 1 -> timepoint when decision boundary is first crossed

trial is ended if decision boundary a is crossed. if not then its not finished. we should find out the percentage 

```python
def euler_maruyama(f_func, g_func, x_0, t, **args):
    y = np.zeros([len(x_0), len(t)])
    dt = t[1] - t[0]
    for i, t_val in enumerate(t):
        if t_val == 0:
            y[:, i] = x_0
        else:
            W = np.random.standard_normal(len(x_0))
            y[:, i] = (y[:, i-1] + np.array(f_func(y[:, i-1], t_val, **args)) * dt \
                       + np.array(g_func(y[:, i-1], t_val,**args)) * np.sqrt(dt) * W)
    return y
```

```python
mus = [-0.1,0.,0.1,0.2,0.5]


def check_ups(a,trials = 200,t=ts):
    up = []
    p =[]
    for mu_1 in mus:
        for i in range(trials):
            ys = euler_maruyama(mu,sigma,[0],t)[0]
            #plt.plot(ts,ys)
            #plt.axhline(1,color ='red')

            #check if boundary is reached
            b = ys>=a
            if b.sum()>0:
                up.append(1)
            else:
                up.append(0)

        p.append(np.mean(up))
    return p
```

```python
p = check_ups(1)

plt.plot(mus,p)
plt.xlabel('$\mu$')
plt.ylabel('percentage of "up"')
```

```python
#resimulation 20 sec
trs = np.arange (0,30,dt)
p = check_ups(1,t=trs)

plt.plot(mus,p)
plt.xlabel('$\mu$')
plt.ylabel('percentage of "up"')
```

```python
#double a
p = check_ups(2)

plt.plot(mus,p)
plt.xlabel('$\mu$')
plt.ylabel('percentage of "up"')
```

```python
#halve a
p = check_ups(0.5)

plt.plot(mus,p)
plt.xlabel('$\mu$')
plt.ylabel('percentage of "up"')
```
