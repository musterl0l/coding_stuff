```python
import numpy as np
from scipy.special import logsumexp
import matplotlib.pyplot as plt
```

#### (1)

```python
#10 uniform numbers 
L = np.random.uniform(0.,0.9,10)
```

#### (2)

```python
#time and lives=instantiations of the agent (you need multiple instantiations for a general statement)

def generate_reward(lever):
    return np.random.choice([0,1], p=[1-L[lever],L[lever]])
```

```python
it = 10000
bandit = np.zeros([10,it])

for a in range(10):
    for i in range (it):
        bandit[a,i] = generate_reward(a)
```

```python
print('difference to reward probabilities:\n\n',np.mean(bandit,axis=1)-L)
```

#### (3)

```python
class Q():
    
    def __init__(self,size,eta=0.01):
        self.values = np.zeros(size)
        self.learnrate = eta
    
    def update(self,i):
        r = generate_reward(i)
        self.values[i] = self.values[i] + self.learnrate*(r-self.values[i])
```

#### (4)

```python
def eps_greedy(qv,eps):
    max_q = np.max(qv.values)
    if (np.random.uniform(0,1) <= (1-eps)):
        max_i, = np.where(qv.values==max_q)
        return max_i.tolist(), max_q
    else:
        rind = np.random.randint(0,9)
        return [rind], max_q
#return lever index always as a list as it is possible that this are more than one
```

#### (5)

```python
#plot expected reward -> he will change it! plot the estimated expected reward = Q-value
eps = 0
it = 1000
fig, ax = plt.subplots()

for j in range(20):
    q = Q(10)
    max_q_list = []
    for i in range(it):
        lever_index, max_q = eps_greedy(q,eps)
        max_q_list.append(max_q)
        for ind in lever_index:
            q.update(ind)
    ax.plot(max_q_list)
plt.show()
```

- the agent does not consistently identify the levers with the largest reward posibilties. Else they would all end up with the same maximum Q-value what they do not
- With the greedy policy the agent can get stuck in a lever with the highest Q-value but not the truely highest reward. It is not exploring much but starts quickly to exploit

#### (6)

```python
eps = [0.0,0.01,0.1,0.5]
it = 1000
lives = 100

fig, ax = plt.subplots()

for e in eps:
    maxQ = np.zeros([lives,it])
    for j in range(lives):
        q = Q(10)
        for i in range(it):
            lever_index, max_q = eps_greedy(q,e)
            maxQ[j,i] = max_q
            for ind in lever_index:
                q.update(ind)
    ax.plot(np.mean(maxQ,axis=0), label=str(e))

plt.legend()
plt.axhline(np.max(L),color='red')
plt.show()
```

- for all epsilons on average the maximum is not reached. for epsilon = 0.0 it is not reached because it gets sometimes stuck with bad q values as analysed in task (5). 
- when epsilon gets higher than 0, there is a uncertainty that is pushing the q-value away again from the maximum value, so with growing epsilon the expected reward can not end up in an optimum policy

#### (7)

```python
#softmax is not numerical stable if input arguments get too large -> use logsumexp
def softmax(qv,beta):
    P = np.exp(beta*qv.values - logsumexp(beta*qv.values))
    maxQ = np.max(qv.values)
    lever = np.random.choice(range(10), p=P)
    return lever,maxQ
```

```python
betas = [1,5,15,50]
it = 5000
lives = 100

fig, ax = plt.subplots()

for b in betas:
    maxQ = np.zeros([lives,it])
    for j in range(lives):
        q = Q(10)
        for i in range(it):
            lever_index, max_q = softmax(q,b)
            maxQ[j,i] = max_q
            q.update(lever_index)
    ax.plot(np.mean(maxQ,axis=0), label=str(b))

plt.legend()
plt.axhline(np.max(L),color='red')
plt.show()
```

- There is a strong improvement compared to the $\epsilon$-greedy policy. for the low beta values we even cross maximum expected reward. the higher beta, the quicker the policy explores but also starts exploiting quicker in suboptimal levers. a beta value of one explores slowly everything and needs some time to reach the maximum. a beta of 5 seems to handle the exploitation/ exploration dilemma best
